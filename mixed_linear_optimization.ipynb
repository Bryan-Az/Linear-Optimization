# Optimization of Mixed Linear Equations 
The goal is to maximize the objective function subject to the constraints.
1. Using a simple linear equation
2. Using a linear equation with mixed variables

## For the linear equations with mixed integers
**Equation**: x + 10y

**Constraints**:

> x+7y <= 17.5

> 0 <= y <= 3.5

> 0 <= y

> x,y integers

```python
def linear_optim_part2():
    """
    constraints: list of lists, each inner list is a constraint of the form [a, b, c]   
    """
    model = ConcreteModel()
    model.x = Var( initialize=-1.2, bounds=(-2, 2) )
    model.y = Var( initialize= 1.0, bounds=(-2, 2) )


    model.limits = ConstraintList()
    model.limits.add(model.x + 7 * model.y <= 17.5)
    model.limits.add(model.x <= 3.5)
    model.limits.add(0 <= model.x)
    model.limits.add(0 <= model.y)

    model.obj = Objective(
    expr= model.x + 10 * model.y,
    sense= maximize )

    SolverFactory('glpk').solve(model, tee=True) and model.display()
```

```python
linear_optim_part2()
```
